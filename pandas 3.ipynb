```python
# another method for filtering
emp1=pd.read_csv("DataSets-master/employees.csv",parse_dates=["Start Date","Last Login Time"])
emp1["Senior Management"]=emp1["Senior Management"].astype(bool)
emp1["Gender"]=emp1["Gender"].astype('category')

emp1.info()
```

```python
# filter db on condition
emp1.head(11)
```

```python
#emp1["Gender"]=="Male" 
emp1[emp1["Gender"]=="Male"].head()
```

```python
emp1[emp1["First Name"]=="Jerry"]
```

```python
emp1[emp1["Senior Management"]!=True].head()
```

```python
emp1[emp1['Salary']>12000].head()
```

```python
# .isin() method
m1=emp1['Team']=="Sales"
m2=emp1["Team"]=="Finance"
m3=emp1["Team"]=="Marketing"

emp1[m1 | m2 | m3].head()
```

```python
m4=emp1["Team"].isin(["Sales","Finance","Marketing"])
emp1[m4].head()
```

```python
# isnull() and notnull()
m5=emp1['Team'].isnull() # extract value which is null exist
emp1[m5].head()
```

```python
m6=emp1['Gender'].notnull() # extract value which is not null exist
emp1[m6].head()
```

```python
# .between() method
m7=emp1['Salary'].between(60000,70000) # extrac salary between 60,000 to 70,000 inclue itself
emp[m7].head()
```

```python
m8=emp1["Last Login Time"].between('9:45AM',"12:00PM") 
emp1[m8].head()
```

```python
# duplicate method()
emp2=emp1.sort_values("First Name")
emp2.head()
```

```python
m9=emp2['First Name'].duplicated() # it take first value as flase and other duplicate values as true
# keep parameter says that first value consider as not duplicate other as duplicate
emp2[m9].head(11)  # extract duplicate value
```

```python
emp2['First Name'].duplicated(keep="last").head()
# keep="last" tells last duplicate value is False other is true
```

```python
emp2['First Name'].duplicated(keep=False)
# it remove first value
```

```python
m10 = ~emp2['First Name'].duplicated(keep="last") # ~ reverse the operation 
emp2[m10].head()
```

```python
# drop_duplicates()
len(emp2)
```

```python
m11=emp2.drop_duplicates(subset=["First Name"],keep="first")
len(m11)
```
